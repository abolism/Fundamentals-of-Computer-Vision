```python


def estimateF(x1,x2):
    # Return the fundamental matrix F (3 by 3), based on two sets of homogeneous 2D points x1 and x2.
    # Input: x1,x2 numpy ndarray (3 by N) containing matching 2D homogeneous points.
    # Output: F numpy ndarray (3 by 3) containing the fundamental matrix.
    N = x1.shape[1]
    A = np.zeros((N, 9))
    for i in range(N):
        A[i] = np.kron(x1[:, i], x2[:, i])
    ignore, ignore, V = np.linalg.svd(A)
    F = V[-1].reshape(3, 3)
    U, S, V = np.linalg.svd(F)
    S[2] = 0
    F = U @ np.diag(S) @ V
    return F

def estimateFnorm(x1,x2):
    # Return the fundamental matrix F (3 by 3), based on two sets of homogeneous 2D points x1 and x2.
    # Input: x1,x2 numpy ndarray (3 by N) containing matching 2D homogeneous points.
    # Output: F numpy ndarray (3 by 3) containing the fundamental matrix based on normalized homogeneous points.
    T1 = np.eye(3)
    T2 = np.eye(3)
    x1_norm = x1 / x1[2]
    x2_norm = x2 / x2[2]
    mean1 = np.mean(x1_norm[:2], axis=1)
    mean2 = np.mean(x2_norm[:2], axis=1)
    scale1 = np.sqrt(2) / np.std(x1_norm[:2])
    scale2 = np.sqrt(2) / np.std(x2_norm[:2])
    T1[:2, 2] = -mean1 * scale1
    T1[:2, :2] = scale1 * np.eye(2)
    T2[:2, 2] = -mean2 * scale2
    T2[:2, :2] = scale2 * np.eye(2)
    x1_norm = T1 @ x1
    x2_norm = T2 @ x2
    F = estimateF(x1_norm, x2_norm)
    F = T2.T @ F @ T1
    return F
```

```python
def vgg_F_from_P(P1, P2):
    # Return the fundamental matrix F (3 by 3), based on two camera parameter arrays.
    # Input: P1, P2 numpy ndarray (3 by 4) containing intrinsic and extrinsic parameters.
    # Output: F numpy ndarray (3 by 3) containing the fundamental matrix.
    X = []
    Y = []
    X.append(P1[[1, 2], :])
    X.append(P1[[2, 0], :])
    X.append(P1[[0, 1], :])
    Y.append(P2[[1, 2], :])
    Y.append(P2[[2, 0], :])
    Y.append(P2[[0, 1], :])
    F = np.zeros([3, 3])

    for i in range(3):
        for j in range(3):
            M = np.concatenate([X[j], Y[i]])
            F[i, j] = np.linalg.det(M)

    # Enforce rank-2 constraint
    U, S, V = np.linalg.svd(F)
    S[2] = 0
    F = U @ np.diag(S) @ V

    # Normalize F
    F /= F[2, 2]

    return F
```
